```python
ed_level_counts = df['EdLevel'].value_counts()

ed_level_counts = ed_level_counts.sort_index()

plt.figure(figsize=(10, 6))
plt.bar(ed_level_counts.index, ed_level_counts.values, color='skyblue')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.title('Distribution of Education Level')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df = df.dropna(subset=['TimeAnswering'])

# Replace 'TimeAnswering' with the actual column name from your DataFrame
count_less_than_15 = df['TimeAnswering'].str.count('Less than 15 minutes a day').sum()
count_15_to_30 = df['TimeAnswering'].str.count('15-30 minutes a day').sum()
count_30_to_60 = df['TimeAnswering'].str.count('30-60 minutes a day').sum()
count_60_to_120 = df['TimeAnswering'].str.count('60-120 minutes a day').sum()
count_over_120 = df['TimeAnswering'].str.count('Over 120 minutes a day').sum()

fig, ax = plt.subplots(figsize=(8, 6))

labels = ['Less than 15 minutes', '15-30 minutes', '30-60 minutes', '60-120 minutes', 'Over 120 minutes']
counts = [count_less_than_15, count_15_to_30, count_30_to_60, count_60_to_120, count_over_120]

colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']

wedges, texts, autotexts = ax.pie(counts, labels=labels, autopct='%1.1f%%',
                                   startangle=90, colors=colors,
                                   textprops=dict(color="w"), shadow=True)

plt.setp(autotexts, size=12, weight="bold")

ax.set_title("Time Answering Distribution")

# Add legend indicating the color for each category
ax.legend(wedges, labels, title="Time Answering", loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

ax.axis('equal')

plt.show()
```
